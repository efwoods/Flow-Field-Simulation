```python
best_K = best_K_seed["K"].values[0]
best_seed = int(best_K_seed["seed"].values[0])
mixture, post = tqdm(common.init(X=X, K=best_K, seed=best_seed), desc="Initializing Mixture...",  ascii="░▒▓█")
mixture, post, ll = tqdm(em.run(X=X, post=post, mixture=mixture), desc="Running EM Algorithm...", ascii="░▒▓█")
X_imputed = em.fill_matrix(X=X, mixture=mixture)
X_imputed_df = pd.DataFrame(X_imputed)
X_imputed_df.describe()
```
